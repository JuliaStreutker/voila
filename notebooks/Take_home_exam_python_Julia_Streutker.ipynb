**Introduction:**

Since the start of the covid 19 pandemic in late 2019 every country has been hit by this global pandemic. However, how countries reacted to the pandemic differed greatly. Some went into a total lockdown and others choose for a partial lockdown. This lead to differences in the new cases that were registered per country. The aim of this study was to determine between the Netherlands, Belgium, France and Germany, which country had the most new cases per month in the year 2021. This study compared mostly the Netherlands and Belgium, and France and  Germany for they are countries located close to each other and are relatively close in amount of residents.  

**Data preperation:**

The columns selected were the iso codes, location, date and the amount of new cases. For the location only the necesarry countries were selected, so the Netherlands, Belgium, Germany and France, and for the date the year 2021 was selected, for it was the only year where covid cases were registered from Januari till December.

```python
import pandas as pd
import altair as alt
import ipywidgets as widgets
covid = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv", parse_dates=['date'])

iso_code_location_date_new_cases = covid[['iso_code', 'location','date', 'new_cases']]
```

```python
year_2021 = iso_code_location_date_new_cases.loc[(covid.date.dt.year == 2021)]
```

```python
countries_to_select = ['Netherlands','Belgium','France','Germany']
year_and_countries = year_2021.loc[year_2021.location.isin(countries_to_select), ]

print (year_and_countries)
```

With this amount of rows, the data could be further explored.

**Explore and clean the data:**

The data was assesed for errors or incorrect data. No errors, missing data or incorrect data were found.  

```python
iso_code_location_date_new_cases.isnull().sum()
```

```python
alt.Chart(year_and_countries).mark_point().encode(
    x='date',
    y='new_cases',
    column='location',
    tooltip='new_cases'
)
```

A chart was made to asses how to display the data of the different countries which were selected. The days are still seperately shown, which proved to not be a good choice, for it didn't have the clarity that was necesarry. This lead to summing the data of each day up per month. By doing so, only one number was show per month, instead of 28-31. This increased the clarity of the graphs.

```python

months_year_2021 = year_and_countries.groupby(pd.Grouper(key='date',freq='M')).sum()
```

```python
final = pd.DataFrame()
for country in countries_to_select:
    final[country] = year_and_countries.loc[year_2021.location.isin([country]), ].groupby(pd.Grouper(key='date',freq='M')).sum()

final
```

The last table shows the amount of new cases per month for each country and proves that the coding was correct.

**Discribe and visualize:**

The final product was visualized making use of a box plot where the different months could be selected.

```python
months = ['Januari','Februari','March','April','May','June','July','August','September','October','November','December']
final.index = months

def visualize_country_data(month):
  
  data_to_plot = pd.DataFrame (final.loc[month])
  data_to_plot.columns = ['new_cases_per_month']
  data_to_plot['location'] = data_to_plot.index
  max_scale_y = final.max().max()
  chart = alt.Chart(data_to_plot).mark_bar().properties(width=150, height= 400,title = 'New cases per month year 2021').configure_title(fontSize=20,
    font='utopia-std, serif',
    anchor='start',
    color='black',
    ).encode(
      x = alt.X('location',axis = alt.Axis(labelFontSize=12, titleFontSize=18, title='Location')),
      y = alt.Y('new_cases_per_month',scale = alt.Scale(domain=[0, max_scale_y]),axis = alt.Axis(labelFontSize=12, titleFontSize=18, title='New cases per month')),
      color= 'location',
      tooltip= 'new_cases_per_month',
  )

  display(chart)

select_month = widgets.Dropdown(options= months)

iout = widgets.interactive_output(visualize_country_data, {'month': select_month})

widgets.HBox([iout, widgets.VBox([select_month])])
```
